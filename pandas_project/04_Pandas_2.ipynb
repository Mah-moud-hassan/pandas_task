### 1. Show value counts for `Student` in the `United Kingdom`

### 2. What's the median sallary `ConvertedComp` for developers in `Germany`,  `Italy`, `Finland`

````python
country_df['ConvertedComp'].median().loc[['Germany' , 'Italy' , 'Finland']]
````

### 3. Show the number of respondents from each country

````python
num_of_respondents=country_df['Respondent'].count()
num_of_respondents
````

### 4. Show the number of respondents that worked with `Android` from each country mentioned in column `PlatformWorkedWith`

````python
number_of_respondents_using_Android=country_df['PlatformWorkedWith'].apply(lambda X:X.str.contains('Android').sum())
number_of_respondents_using_Android
````

### 5. the function **`pandas.concat()`** to add these two series together in a new DataFrame

````python
new_num_of_respondents= pd.concat([num_of_respondents, number_of_respondents_using_Android],axis=1,ignore_index=True)
````

````python
new_num_of_respondents = pd.DataFrame(new_num_of_respondents)
new_num_of_respondents.columns = [['num_of_respondents','number_of_respondents_using_Android']]
````

````python
new_num_of_respondents.head(5)
````

### 6. Add a new column contains the percentage of people that worked with `Android` from each country

````python
number_of_respondents_all_plats=country_df['PlatformWorkedWith'].count()
number_of_respondents_all_plats
````

````python
number_of_respondents_using_Android=country_df['PlatformWorkedWith'].apply(lambda X:X.str.contains('Android').sum())
number_of_respondents_using_Android
````

````python
percentage_of_Android = round(((number_of_respondents_using_Android)/(number_of_respondents_all_plats))*100,2)
````

````python
percentage_of_Android
````

````python
new_num_of_respondents['percentage_of_Android'] = percentage_of_Android
````

````python
new_num_of_respondents.head(5)
````
